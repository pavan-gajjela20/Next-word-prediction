```python
corpus = ["The quick brown fox jumps over the lazy dog.A journey of a thousand miles begins with a single step.""To be or not to be, that is the question.All that glitters is not gold.Every cloud has a silver lining."
"In the middle of difficulty lies opportunity.The early bird catches the worm.Actions speak louder than words.Better late than never.Fortune favors the bold."]

corpus
```

```python
for y in corpus:
    print(y)
    print("-"*29)
```

```python
import tensorflow as ts
import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tk = Tokenizer()
```

```python
df = tk.fit_on_texts(corpus)
```

```python
print(df)
```

```python
len(tk.word_index)
```

```python
df = tk.texts_to_sequences(corpus)
print(df)
```

```python
input = []
output = []
for token in df:
    for index in range(len(token)-1):
        input.append(token[:index+1])
        output.append(token[index+1])
        
```

```python
input
```

```python
output
```

```python
len(input)
```

```python
len(output)
```

```python
from tensorflow.keras.utils import pad_sequences, to_categorical
```

```python
feature_variable=pad_sequences(input,padding='post')
```

```python
pad_sequences(input).shape
```

```python
feature_variable.shape
```

```python
input
```

```python
class_variable = to_categorical(output,num_classes=66+1)
class_variable
```

```python
class_variable.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, InputLayer
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(67,)))
model.add(Embedding(52,2))
model.add(SimpleRNN(10, return_sequences=False))
model.add(Dense(67,activation="softmax"))
```

```python
model.compile(
    optimizer='adam',
    loss="categorical_crossentropy",
    metrics=["accuracy"])
```

```python
model.fit(feature_variable,class_variable,epochs=10)
```

```python
model.fit(feature_variable,class_variable,epochs=100)

## the more epochs then model will learn more information
```

```python
import numpy as np
```

```python
x = ['quick']
np.argmax(model.predict(pad_sequences(tk.texts_to_sequences(x),maxlen=67)))
```
